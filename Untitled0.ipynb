Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten
```

 Load Fashion MNIST dataset


```python
(x_train, y_train), (_, _) = fashion_mnist.load_data()
x_train = x_train.astype('float32') / 255.0
x_train = np.expand_dims(x_train, -1)
```

Define CNN for feature extraction

```python
input_img = Input(shape=(28, 28, 1))
x = Conv2D(32, (3, 3), activation='relu')(input_img)
x = MaxPooling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu')(x)
x = MaxPooling2D((2, 2))(x)
x = Flatten()(x)

feature_extractor = Model(inputs=input_img, outputs=x)
features = feature_extractor.predict(x_train, batch_size=256, verbose=1)
```

Apply K-Means clustering


```python
n_clusters = 10  # Fashion MNIST has 10 categories
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(features)
```

Reduce dimensions with PCA for visualization

```python
pca = PCA(n_components=2)
pca_features = pca.fit_transform(features)


```

Visualize clusters

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x=pca_features[:, 0],
    y=pca_features[:, 1],
    hue=clusters,
    palette='tab10',
    legend='full'
)
plt.title("PCA Visualization of CNN + KMeans Clusters")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.show()
```

 Show example images from each cluster

```python
def show_cluster_examples(x, clusters, cluster_id, n=10):
    idxs = np.where(clusters == cluster_id)[0][:n]
    plt.figure(figsize=(n, 2))
    for i, idx in enumerate(idxs):
        plt.subplot(1, n, i+1)
        plt.imshow(x[idx].squeeze(), cmap='gray')
        plt.axis('off')
    plt.suptitle(f'Cluster {cluster_id}')
    plt.show()
```

Display samples from each cluster

```python
for i in range(n_clusters):
    show_cluster_examples(x_train, clusters, i)
```
